```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set(style='darkgrid')

mpl.rc('font', family='Arial')
```

```python
df = pd.read_csv('Automobile_data.csv')
df.head(7)
```

```python
sns.scatterplot(data=df, x='wheel-base', y='length')
```

```python
sns.scatterplot(data=df, x='wheel-base', y='length', hue="company")  #overlay
```

```python
sns.lmplot(data=df, x='wheel-base', y='length' )
```

```python
sns.jointplot(data=df, x='wheel-base', y='length')
```

```python
sns.pairplot(df[['wheel-base','length','company']],hue='company')
```

```python
df.corr()
```

```python
sns.lmplot(data=df, x='price', y='length' )
```

## Seaborn and Bokeh (Web )

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(data=tips, x='total_bill', y='tip')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', palette='winter')
```

```python
tips
sns.catplot(x='day',y='total_bill', data=tips)
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='violin')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='bar')
```
